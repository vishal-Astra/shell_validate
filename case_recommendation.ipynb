```python
preparation = fix_case
preparation.sort_values("isbn")
```

```python
preparation = preparation.drop_duplicates("isbn")
preparation
```

```python
data = preparation.drop(columns=["case_count"])
data.sample(5)
```

```python

tf = TfidfVectorizer()


tf.fit(data["case"]) 
 

tf.get_feature_names() 
```

```python

tfidf_matrix = tf.fit_transform(data["case"])

tfidf_matrix.shape 
```

```python

tfidf_matrix.todense()
```

```python
 
#pd.DataFrame(
 #   tfidf_matrix.todense(), 
  #  columns=tf.get_feature_names(),
   # index=data.verdict
#).sample(22, axis=1).sample(10, axis=0)
```

```python

cosine_sim = cosine_similarity(tfidf_matrix) 
cosine_sim
```

```python

cosine_sim_df = pd.DataFrame(cosine_sim, index=data['verdict'], columns=data['verdict'])
print('Shape:', cosine_sim_df.shape)

cosine_sim_df.sample(5, axis=1).sample(10, axis=0)
```

```python
def case_recommendations(case_verdict, similarity_data=cosine_sim_df, items=data[['verdict', 'case']], k=5):


    index = similarity_data.loc[:,case_verdict].to_numpy().argpartition(
        range(-1, -k, -1))
    

    closest = similarity_data.columns[index[-1:-(k+2):-1]]
    

    closest = closest.drop(case_verdict, errors='ignore')
 
    return pd.DataFrame(closest).merge(items).head(k)
```
